**1.1** Prove that $E[aX + b] = aE[X] + b$

$$
E[aX + b] = \sum_{x} (aX + b)p(X) \\
\Rightarrow a \sum_{x} X p(X) + b \sum_{x} p(X) \\
\text{We know that } \sum_{x} p(X) = 1 \\
\text{Thus, } E[aX + b] = a\sum_{x} X p(X) + b \\
\text{And } \sum_{x} X p(X) = E[X] \\
\text{Therefore, the equation becomes } aE[X] + b \\
\text{Hence, } E[aX + b] = aE[X] + b
$$


**1.2** Prove that $Var(X) = E[X^2] - E[X]^2$.

$$
\begin{aligned}
\mathrm{Var}(X) &= E[(X - \mu)^2] \\
\mathrm{Var}(X) &= E[X^2 - 2\mu X + \mu^2] \\
\mathrm{Var}(X) &= E[X^2] - E[\mu(2X - \mu)] \\
\mathrm{Var}(X) &= E[X^2] - \mu \cdot E[2X - \mu] \\
\mathrm{Var}(X) &= E[X^2] - \mu \cdot (2E[X] - E[\mu]) \\
\mathrm{Var}(X) &= E[X^2] - \mu \cdot (2\mu - \mu) \\
\mathrm{Var}(X) &= E[X^2] - \mu^2 \\
\mathrm{Var}(X) &= E[X^2] - (E[X])^2
\end{aligned}
$$

##### **1.3** Prove that $Var(X+Y) = Var(X) + Var(Y) + 2Cov(X,Y)$

$$
\begin{aligned}
\text{We know that, } \mathrm{Var}(Z) &= E[Z^2] - (E[Z])^2 \\
\text{Thus, } \mathrm{Var}(X+Y) &= E[((X+Y) - E[X+Y])^2] \\
\text{As } E[X+Y] &= E[X] + E[Y] \\
\text{Thus, } \mathrm{Var}(X+Y) &= E[((X+Y) - E[X] - E[Y])^2] \\
\text{Rearranging, } \mathrm{Var}(X+Y) &= E[((X - E[X]) + (Y - E[Y]))^2] \\
\text{Expanding, } \mathrm{Var}(X+Y) &= E[(X - E[X])^2 + (Y - E[Y])^2 + 2(X - E[X])(Y - E[Y])] \\
\mathrm{Var}(X+Y) &= E[(X - E[X])^2] + E[(Y - E[Y])^2] + 2E[(X - E[X])(Y - E[Y])] \\
\mathrm{Var}(X+Y) &= \mathrm{Var}(X) + \mathrm{Var}(Y) + 2\,\mathrm{Cov}(X,Y) \\
\end{aligned}
$$

Prove that $E[X] = \frac{a+b}{2}$ and $Var(X) = \frac{(b-a)^2}{12}$.

First, finding the expectation of X.

$$
\begin{aligned}
E[X] &= \int_a^b x f(x)\, dx \\
E[X] &= \int_a^b x \cdot \frac{1}{b-a}\, dx \\
E[X] &= \frac{1}{b-a} \int_a^b x\, dx \\
E[X] &= \frac{1}{b-a} \cdot \left[ \frac{x^2}{2} \right]_a^b \\
E[X] &= \frac{1}{b-a} \cdot \left[ \frac{b^2}{2} - \frac{a^2}{2} \right] \\
E[X] &= \frac{1}{2(b-a)} \cdot \left[b^2 - a^2 \right]\\
E[X] &= \frac{1}{2(b-a)} \cdot (b-a)(b+a) \\
E[X] &= \frac{a+b}{2} \\
\end{aligned}
$$


Now, finding the variance of X

$$
\begin{aligned}
\text{We know that, } V[X] &= E[X^2] - (E[X])^2 \\
\text{Finding } E[X^2]: \\
E[X^2] &= \int_a^b x^2 f(x)\, dx \\
E[X^2] &= \int_a^b x^2 \cdot \frac{1}{b-a}\, dx \\
E[X^2] &= \frac{1}{b-a} \left[ \frac{x^3}{3} \Big|_a^b \right] \\
E[X^2] &= \frac{1}{b-a} \left( \frac{b^3}{3} - \frac{a^3}{3} \right) \\
E[X^2] &= \frac{b^3 - a^3}{3(b-a)} \\
E[X^2] &= \frac{1}{3}(b^2 + ab + a^2) \\[6pt]
\text{Plugging into the variance formula:} \\
V[X] &= \frac{1}{3}(b^2 + ab + a^2) - \left(\frac{a+b}{2}\right)^2 \\
V[X] &= \frac{4(b^2 + ab + a^2) - 3(a^2 + 2ab + b^2)}{12} \\
V[X] &= \frac{b^2 - 2ab + a^2}{12} \\
V[X] &= \frac{(b-a)^2}{12}
\end{aligned}

```python
# importing the necessary libraries
import rasterio
import glob
import os
import numpy as np
import matplotlib.pyplot as plt
from rasterio.merge import merge
from rasterio.warp import calculate_default_transform, reproject, Resampling
from rasterio.crs import CRS
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.cm as cm
```

```python
# defining the target bounding box for Durham
# when selecting the HLS data, I did a polygon selection so I have more data than needed
target_bounds = [-79.0, 35.9, -78.8, 36.1]

# reprojecting files into a Plate Carree projection
target_crs = CRS.from_epsg(4326)

# 1 arcsecond 
target_res = 1/3600

data_dir = './Q2_data'

# required bands for RGB (B4, B3, B2) and NDVI (B8, B4)
required_bands = {'B02': 'Blue',
                 'B03': 'Green',
                 'B04': 'Red',
                 'B08': 'NIR'}

# output directory for processed files
output_dir = './Q2_processed_data'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

processed_paths = {}
```

```python
"""
I was facing an issue before where the files I was reading were corrupted
This code chunk is supposed to help identify if the files are corrupted or not
"""
corrupted_files = []
good_files = []

for filename in os.listdir(data_dir):
    if filename.endswith('.tif'):
        filepath = os.path.join(data_dir, filename)
        try:
            with rasterio.open(filepath) as src:
                # reading the entire file to check for corruption
                # reading the full band instead of just one pixel
                _ = src.read(1)
            good_files.append(filename)
            print(f"✓ {filename}")
        except Exception as e:
            corrupted_files.append(filename)
            print(f"✗ {filename}: {str(e)[:80]}")

print(f"\nGood files: {len(good_files)}")
print(f"Corrupted files: {len(corrupted_files)}")
if corrupted_files:
    print("\nCorrupted files:")
    for f in corrupted_files:
        print(f"  - {f}")
```

```python
# processing each band 
for band_id, band_name in required_bands.items():
    print(f"\nProcessing Band {band_id} ({band_name})...")
    
    # finding all the files for a specific band
    search_path = os.path.join(data_dir, f'*{band_id}.tif')
    band_files = glob.glob(search_path)
    print(f"  Found {len(band_files)} files")
    
    src_datasets = []
    
    for fp in band_files:
        ds = rasterio.open(fp)
        # test reading the entire dataset before adding it to the list
        test_data = ds.read(1)
        print(f"  Successfully read {os.path.basename(fp)}")
        src_datasets.append(ds)
    
    if not src_datasets:
        print(f"Error: No valid files found for Band {band_id}. Skipping.")
        continue
    
    print(f"  Merging {len(src_datasets)} tiles...")
    
    mosaic, out_trans = merge(src_datasets)
    print(f"  Merge successful!")
    
    # getting the source profile from the first dataset
    src_profile = src_datasets[0].profile
    src_crs = src_datasets[0].crs
    
    # calculating the transform for reprojection to the target CRS and resolution
    # using the bounds of the mosaic
    left, bottom, right, top = rasterio.transform.array_bounds(
        mosaic.shape[1], mosaic.shape[2], out_trans
    )

    transform, width, height = calculate_default_transform(
        src_crs, target_crs,
        mosaic.shape[2], mosaic.shape[1],
        left, bottom, right, top,
        resolution = (target_res, target_res)
    )

    # creating an output array for the reprojection
    reprojected = np.zeros((1, height, width), dtype = mosaic.dtype)

    # reprojecting the mosaic
    print(f'Reprojecting to EPSG:4326 at 1 arcsec resolution..')
    reproject(
        source = mosaic,
        destination = reprojected,
        src_transform = out_trans,
        src_crs = src_crs,
        dst_transform = transform,
        dst_crs = target_crs,
        resampling = Resampling.bilinear
    )
    
    # subsetting to Durham bounding box
    print(f'Subsetting to Durham bounds...')

    # calculating the ixel coordinates for the bounding box
    inv_transform = ~transform
    col_min, row_min = inv_transform * (target_bounds[0], target_bounds[3])
    col_max, row_max = inv_transform * (target_bounds[2], target_bounds[1])

    col_min, col_max = int(col_min), int(col_max)
    row_min, row_max = int(row_min), int(row_max)

    # extracting the subset
    subset = reprojected[0, row_min:row_max, col_min:col_max]

    # updating transform for the subset
    subset_transform = rasterio.transform.from_bounds(
        target_bounds[0], target_bounds[1],
        target_bounds[2], target_bounds[3],
        subset.shape[1], subset.shape[0]
    )
    
    # saving the processed band
    output_path = os.path.join(output_dir, f'{band_id}_processed.tif')
    with rasterio.open(
        output_path, 'w',
        driver = 'GTiff',
        height = subset.shape[0],
        width = subset.shape[1],
        count = 1,
        dtype = subset.dtype,
        crs = target_crs,
        transform = subset_transform,
        nodata = 0
    ) as dst:
        dst.write(subset, 1)
    
    processed_paths[band_id] = output_path
    print(f'Saved to {output_path}')

    # close all source datasets
    for ds in src_datasets:
        ds.close()
```

```python
# loading the processed bands for visualization
print('Loading the processed bands for visualization...')
bands = {}
for band_id in ['B02', 'B03', 'B04', 'B08']:
    with rasterio.open(processed_paths[band_id]) as src:
        bands[band_id] = src.read(1).astype(float)

# creating the RGB composite by stacking the B04 (red), B03 (green) and B02 (blue) bands
rgb = np.stack([bands['B04'], bands['B03'], bands['B02']], axis = 2)

# normalize to a 0 to 1 range
# replace the zeros with NaN for better visualization
for i in range(3):
    band = rgb[:, :, i]
    band[band == 0] = np.nan
    valid_data = band[~np.isnan(band)]
    if len(valid_data) > 0:
        # normalize using percentile stretch for a better contrast
        p2, p98 = np.percentile(valid_data, [2, 98])
        band = np.clip(band, p2, p98)
        band = (band - p2) / (p98 - p2)
        rgb[:, :, i] = band
```

```python
# calculating NDVI
print('Calculating NDVI...')
nir = bands['B08'].astype(float)
red = bands['B04'].astype(float)

# masking out the nodata values
nir_masked = np.where(nir == 0, np.nan, nir)
red_masked = np.where(red == 0, np.nan, red)

# calculating NDVI
ndvi = (nir_masked - red_masked) / (nir_masked + red_masked)
print(f'NDVI range: {np.nanmin(ndvi):.3f} to {np.nanmax(ndvi):.3f}')
```

```python
# creating the plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# plotting RGB
ax1.imshow(rgb)
ax1.set_title('RGB Composite - Durham, NC\nSentinel-2 (September 10, 2021)',
             fontsize = 14, fontweight = 'bold')
ax1.set_xlabel('Longitude', fontsize = 12)
ax1.set_ylabel('Latitude', fontsize = 12)
ax1.grid(True, alpha = 0.4)

# setting up the colormap for NDVI
cmap = cm.RdYlGn
# setting the NaN values to gray
cmap.set_bad(color='gray')

# plotting NDVI
ndvi_min = np.nanpercentile(ndvi, 1)
ndvi_max = np.nanpercentile(ndvi, 99) 
im = ax2.imshow(ndvi, cmap=cmap, vmin=ndvi_min, vmax=ndvi_max, interpolation = 'nearest')
ax2.set_title('NDVI - Durham, NC\nSentinel-2 (September 10, 2021)',
             fontsize = 14, fontweight = 'bold')
ax2.set_xlabel('Longitude', fontsize = 12)
ax2.set_ylabel('Latitude', fontsize = 12)
ax2.grid(True, alpha = 0.4)

# adding a colorbar for NDVI
cbar = plt.colorbar(im, ax = ax2, fraction = 0.046, pad = 0.04)
cbar.set_label('NDVI', fontsize = 12)

plt.tight_layout()
plt.show()
```

# 3. Calculating the correct spatial mean (20 pts)

A simple spatial mean of a regular lat/long grid (Plate Carree) can lead to large discrepancies in the estimated spatial mean. To assess this impact, we are going to use the variable `t2m` in the annual ERA-Interim database (`/data/era-interim/era_interim_monthly_197901_201512_upscaled.nc_ann`). Start by assembling an array of actual area per pixel and then compute for each year the area weighted mean. Then compare the resulting time series to that resulting from using the equally weighted spatial mean (i.e., naive approach). Provide an explanation of your results.

```python
# importing the necessary libraries
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

# defining the file path
file_path = "/data/era-interim/era_interim_monthly_197901_201512_upscaled.nc_ann"

# earth's radius in meters
r = 6.371e6

ds = xr.open_dataset(file_path)
t2m = ds["t2m"]
```

```python
# converting the latitude coordinates to radians
lat_rad = np.deg2rad(t2m.lat)
lon_rad = np.deg2rad(t2m.lon)

# calculating the difference in longitude which would be constant for a regular grid
dlon = np.abs(lon_rad.diff('lon')[0].item())

# getting the latitude midpoints, here boundaries
lat_bounds = (t2m.lat.rolling(lat=2).mean()
             .shift(lat = -1)
             .bfill(dim = 'lat')
             .ffill(dim = 'lat'))

# determining the full boundaries assuming uniform spacing at the poles
dlat = np.abs(t2m.lat.diff('lat')[0].item())

# boundary coordinates for the South and North Pole
lat_min_pole = t2m.lat.min().item() - dlat/2
lat_max_pole = t2m.lat.max().item() + dlat/2

# constructing the full boundaries array
lat_boundaries = np.concatenate([
    [lat_min_pole],
    (t2m.lat.data[:-1] + t2m.lat.data[1:]) /2,
    [lat_max_pole]
])
lat_boundaries_rad = np.deg2rad(lat_boundaries)

# calculating an area array
area_weights = (
    r**2 * dlon * (np.sin(lat_boundaries_rad[1:]) - np.sin(lat_boundaries_rad[:-1])))

# converting this area array from a 1D array to a 1D array which matches the latitude dimension
area_da_1d = xr.DataArray(
    area_weights,
    coords = {'lat': t2m.lat.data},
    dims = ['lat'],
    name = 'area_weights_1d')
```

```python
# creating a 2D area array (lat,lon) by expanding the longitude dimension
area_da_2d = area_da_1d.expand_dims(lon=t2m.lon.data, axis = 1)

# multiplying T2M by the area weights
t2m_weighted = t2m * area_da_2d

# summing the weighted temperature over all the spatial dimensions
sum_t2m_weighted = t2m_weighted.sum(dim=['lat','lon'])

# calculating the total area
total_area = area_da_2d.sum(dim=['lat', 'lon'])

# dividing the sum of the weighted termperature by the total area
area_weighted_mean = sum_t2m_weighted / total_area
area_weighted_mean.name = 'area_weighted_mean'
```

```python
# computing the naive spatial mean which is equally weighted
naive_mean = t2m.mean(dim = ['lat', 'lon'])
naive_mean.name = 'naive_mean'
units = t2m.units
```

```python
# calculating the difference
difference = naive_mean - area_weighted_mean
mean_difference = difference.mean().item()
```

```python
# plotting the time series
plt.figure(figsize = (10,6))
area_weighted_mean.plot(label = 'Area-Weighted Mean', color = 'r', linewidth = 2)
naive_mean.plot(label='Naive (Equally-Weighted) Mean', color = 'b', linestyle = '--')
plt.title('Comparison of Spatial Mean Global T2M')
plt.ylabel(f'Temperature (°C)')
plt.xlabel('Year')
plt.legend()
plt.grid(True, alpha = 0.4)
plt.show()
```

```python
# printing the results
print(f'Mean Area-Weighted T2M over the period: {area_weighted_mean.mean().item():.3f} °C')
print(f'Mean Naive T2M over the period: {naive_mean.mean().item():.3f} °C')
print(f'Mean difference between the Naive and the Area-Weighted: {mean_difference:.3f} °C')
```

> The results show that the Naive (Equally Weighted) mean is consistently lower than the Area-Weighted Mean across the entire 1979-2016 period, with an average difference of approximately -8.96°C. This highlights the error that occurs when calculating the mean from a regular latitude/longitude or the Plate Carree grid without accounting for the geometry of the earth.
>
> The naive mean assigns an equal weight to every pixel regardless of the size. Thus, small-area pixels found in the polar regions are given the same influence on the mean as the fewer, large-area pixels found near the equator. As the temperature is much colder at the polar regions than in the tropics and these are assigned equal weights, we can see that the naive mean is much lower than the true mean. The area-weighted method corrects for this distortion by weighting each pixel's temperature by it's true physical area.

Given the shapefile of county boundaries (`/data/HW/HW2/NCDOT_County_Boundaries.shp`) and the USGS National Elevation Dataset (`/data/NED/NED.vrt`), compute the following statistics: the spatial mean, the spatial standard deviation, and the coefficient of variation of elevation per NC county.
```python
# importing the necessary libraries
import geopandas as gpd
import rasterio
from rasterio.mask import mask
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import mapping
from shapely.ops import transform

# loading the NC county boundaries shapefile
counties = gpd.read_file('/data/HW/HW2/NCDOT_County_Boundaries.shp')

# loading the elevation data
ned_path = '/data/NED/NED.vrt'
```

```python
print("COUNTIES SHAPEFILE INFO")
print(f"CRS: {counties.crs}")
print(f"Bounds: {counties.total_bounds}")
print(f"Number of counties: {len(counties)}")
print(f"\nFirst few county names:")
print(counties['CountyName'].head(10))
```

```python
# reprojecting counties to EPSG:4326 with proper axis handling
counties_reprojected = counties.to_crs('EPSG:4326')
print(f'Reprojected bounds: {counties_reprojected.total_bounds}')
print('Expected NC bounds: [lon_min, lat_min, lon_max, lat_max] = [-84.3, 33.8, -75.4, 36.6]')
```

```python
# when projecting, the x and y coordinates got swapped. reswapping 
minx, miny, maxx, maxy = counties_reprojected.total_bounds

counties_reprojected['geometry'] = counties_reprojected['geometry'].apply(
    lambda geom: transform(lambda x, y: (y, x), geom)
)
print(f'After swap: {counties_reprojected.total_bounds}')
```

```python
# initializing the lists to store results
county_names = []
mean_elevations = []
std_elevations = []
coef_elevations = []
```

```python
counties.keys()
```

```python
# processing each county
print('Processing counties...')
for idx, county in counties_reprojected.iterrows():
    county_name = county['CountyName']
    
    # extracting the geometry for the county
    geom = [mapping(county.geometry)]
    
    try:
        # opening the raster and extracting the data for this county
        with rasterio.open(ned_path) as src:
            # masking the raster with the county boundary
            out_image, out_transform = mask(src, geom, crop=True, nodata=np.nan)
            
            # extracting the elevation values
            elevation = out_image[0]
            
            # removing the nodata values
            valid_elevation = elevation[~np.isnan(elevation)]
            
            if len(valid_elevation) > 0:
                # calculating the statistics
                mean_elevation = np.mean(valid_elevation)
                std_elevation = np.std(valid_elevation)
                
                # coefficient of variation
                if mean_elevation > 0:
                    coef_elevation = std_elevation / mean_elevation
                else:
                    coef_elevation = np.nan
                
                # ADD COUNTY NAME HERE - only if successful
                county_names.append(county_name)
                mean_elevations.append(mean_elevation)
                std_elevations.append(std_elevation)
                coef_elevations.append(coef_elevation)
            else:
                # there is no valid data for this county
                county_names.append(county_name)  # ADD HERE TOO
                mean_elevations.append(np.nan)
                std_elevations.append(np.nan)
                coef_elevations.append(np.nan)
        
        print(f'✓ Processed: {county_name}')
        
    except Exception as e:
        # If any error occurs, still append values to keep arrays same length
        print(f'✗ Error processing {county_name}: {e}')
        county_names.append(county_name)  # ADD HERE AS WELL
        mean_elevations.append(np.nan)
        std_elevations.append(np.nan)
        coef_elevations.append(np.nan)
```

# 5. Flying from LAX to LHR (20 pts)

Create a rotating globe that tracks an actual flight from Los Angeles to London that happened between 2020/09/17-2020/09/18. The file `/data/HW/HW2/lax2lhr_20200917.pck` contains the GPS location of the flight at discrete timestamps during the flight. The goal here is to create a movie where each frame builds on the previous one to build out the time evolving flight path. The flight path should be drawn as lines joined at nodes. Each frame update should correspond to a new timestamp. Furthermore, for each timestamp, the globe should be centered on the timestamp's corresponding longitude. Finally, add the time evolving rainfall maps that was forecasted during each timestamp. For this part, you will use a 9/17/2020 forecast from NOAA's Global Forecasting System (GFS) which you can find in  `/data/HW/HW2/gfs20200917_gfs_0p25_1hr_00z.nc`. The variable of interest in that dataset will be `pratesfc`. Note that there are many ways to do this problem so feel free to explore your own ideas. However, for reference, below is an example of what the last frame might look like. We look forward to seeing what you come up with! 


```python
# importing the required libraries
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import pickle
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# defining the data paths
flight_path = '/data/HW/HW2/lax2lhr_20200917.pck'
gfs_path = '/data/HW/HW2/gfs20200917_gfs_0p25_1hr_00z.nc'
```

```python
# constants for plotting
rain_colormap = 'Blues'
rain_levels = np.linspace(1e-6, 1e-4, 10)
line_color = 'blue'
line_width = 2
marker_color = 'darkblue'

# loading the flight data
with open(flight_path, 'rb') as f:
    flight_data = pickle.load(f)
    
# checking the columns in the flight data so we can identify the columns properly
flight_data.keys()
```

```python
# extracting the coordinates and timestamps
flight_lons = flight_data['lon']
flight_lats = flight_data['lat']
flight_times = flight_data['date']
```

```python
# loading the GFS rainfall forecast data
gfs_ds = xr.open_dataset(gfs_path)

# checking the columns in the flight data so we can identify the columns properly
gfs_ds.keys()
```

```python
rainfall = gfs_ds['pratesfc']

num_frames = len(flight_lons)
```

```python
# creating the figure
fig = plt.figure(figsize=(10, 10))

def init():
    return []

def update(frame):
    # get the current position
    current_lon = flight_lons[frame]
    current_lat = flight_lats[frame]
    current_time = np.datetime64(flight_times[frame])
    
    # clearing the figure and recreating axes with new projection center
    fig.clear()
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.Orthographic(
        central_longitude=current_lon,
        central_latitude=current_lat
    ))
    ax.set_global()
    
    # ADD THIS LINE: Natural Earth background image
    ax.stock_img()
    
    # drawing background features on top
    ax.add_feature(cfeature.COASTLINE, linewidth=0.5, edgecolor='black', zorder=2)
    ax.add_feature(cfeature.BORDERS, linestyle=':', edgecolor='gray', zorder=2)
    
    # plotting the rainfall forecast
    time_diff = np.abs(gfs_ds.time - current_time)
    closest_idx = time_diff.argmin().item()
    current_rain = rainfall.isel(time=closest_idx)
    
    # fixing the longitudinal range of the rainfall
    current_rain = current_rain.assign_coords(
    lon=(((current_rain.lon + 180) % 360) - 180))
    
    
    # only plot the rainfall where it exists
    rain_masked = current_rain.where(current_rain != 0)
    
    im = ax.pcolormesh(
        rain_masked.lon,
        rain_masked.lat,
        rain_masked,
        transform=ccrs.PlateCarree(),
        cmap=rain_colormap,
        vmin=0,
        vmax=np.nanpercentile(current_rain.values, 99),
        alpha=0.5,
        zorder=3,
        shading='auto'
    )
    
    # plotting the accumulated flight path up to the current frame
    if frame > 0:
        path_lons = flight_lons[:frame+1]
        path_lats = flight_lats[:frame+1]
        
        ax.plot(
            path_lons, path_lats,
            color=line_color,
            linewidth=line_width,
            transform=ccrs.Geodetic(),
            zorder=4
        )
        
    # plotting the current position marker
    ax.plot(
        current_lon, current_lat,
        color=marker_color,
        marker='o',
        markersize=8,
        transform=ccrs.Geodetic(),
        zorder=5
    )
    
    # updating the title with the timestamp
    time_str = np.datetime_as_string(current_time, unit='s').replace('T', ' ')
    ax.set_title(f'Flight LAX to LHR\n{time_str}', fontsize=14, fontweight='bold')
    
    return []
```

```python
# creating the animation
anim = FuncAnimation(
    fig,
    update,
    frames=num_frames,
    init_func=init,
    interval=200,
    repeat=True,
    blit=False
)

# displaying in the workbook itself
HTML(anim.to_html5_video())
```
