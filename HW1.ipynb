## 3. Calculating the Number of Daylight Hours (10 points)

There are a few different models, with varying degrees of accuracy, to estimate the number of daylight hours at a given time and place for application in agriculture, ecology, and earth system modeling. 

A relatively simplistic one from Brock (1981) is as follows:

First, the declination of the sun is given by:

$\delta = .40928 \sin{(2\pi \frac{283+J}{365})}$

where $J$ is the day of the year (number of days since January 1st). The sunset/sunrise hour angle is then computed as 

$\beta = \arccos(-\tan{\phi}\tan{\delta})$

Where $\phi$ is the latitude in radians. Finally the daylength, in hours, is calculated

$D = 24\frac{\beta}{\pi}$

**3.1** Create a function named `daylength` that computes number of daylight hours at a given time and place. Your input parameters should be `lat` and `day` (for day of the year). Keep in mind that latitude is usually given in degrees but the equation requires latitude in radians.

````python
def daylength(lat, day):
    """
    Calculate the daylength (in hours) for a given latitude and day of the year.
    Parameters:
    lat : float or np.ndarray
        Latitude in degrees. Positive for Northern Hemisphere, negative for Southern Hemisphere.
    day : int
        Day of the year (1 to 365).
    Returns:
    float or np.ndarray
        Daylength in hours.
    """
    # converting latitude to radians
    phi = np.radians(lat)

    # solar declination
    delta = 0.40928 * np.sin(2 * np.pi * (283 + day) / 365)

    # sunset/sunrise hour angle
    beta = np.arccos(-np.tan(phi) * np.tan(delta))

    # daylength
    D = 24 * beta / np.pi

    return D
````

````python
# function to read AAIGrid file
def read_aaigrid(path):
    header = {}
    with open(path) as f:
        for _ in range(6):
            key, val = f.readline().strip().split()
            header[key.lower()] = float(val)
        data = np.loadtxt(path, skiprows = 6)
        nodata = header.get("nodata_value", -9999)
        data[data == nodata] = np.nan
        return data, header
````

Using the file /data/era-interim/era_interim_monthly_197901_201512_upscaled.nc:

The following example will show you how to make a movie in a Jupyter Notebook: https://github.com/BindiChen/machine-learning/blob/main/data-visualization/0001-matplotlib-animation/matplotlib-animation-notebook.ipynb. [Hint: You will need to use ax.imshow() instead of ax.line()]. 


````python
import matplotlib.animation as animation

# loading the dataset
ds = nc.Dataset("/data/era-interim/era_interim_monthly_197901_201512_upscaled.nc")

# exploring variables 
print(ds.variables.keys())
````

````python
# extracting the keys from the dataset
precip = ds.variables['precip'][:]
time = ds.variables['time'][:]
lat = ds.variables['lat'][:]
lon = ds.variables['lon'][:]
````

````python
# converting time units
time_units = ds.variables['time'].units
calendar = ds.variables['time'].calendar if hasattr(ds.variables['time'], 'calendar') else 'standard'
dates = nc.num2date(time, units=time_units, calendar=calendar)
````

````python
# reshaping it into [years, months, lat, lon]
# currently the shape is [time, lat, lon]
nlat, nlon = len(lat), len(lon)
nyears = len(np.unique([d.year for d in dates]))
nmonths = 12

# indexing the months
months = np.array([d.month for d in dates])
````

````python
# computing the monthly climatology from January to December
monthly_mean = np.stack([
    precip[months == m, :, :].mean(axis=0)
    for m in range(1, 13)
])
````

````python
# setting up the plot
fig, ax = plt.subplots(figsize = (8,4))
cax = ax.imshow(monthly_mean[0, :, :], origin = "lower", extent = [lon.min(), lat.min(), lon.max(), lat.max()])
ax.set_title("Global Precipitation - Month 1 (Jan)")
fig.colorbar(cax, ax=ax, label= "Precipitation (mm/day or kg/m2/s depending on units)")
````

````python
# setting up the animation
fig, ax = plt.subplots(figsize=(8, 4))

# create the image once
im = ax.imshow(monthly_mean[0, :, :],
               origin="lower",
               extent=[lon.min(), lon.max(), lat.min(), lat.max()],
               cmap="Blues")
fig.colorbar(im, ax=ax, label="Precipitation")
title = ax.set_title("Global Precipitation - January")

# update function
def update(frame):
    im.set_data(monthly_mean[frame, :, :])        # update data
    title.set_text(f"Global Precipitation - Month {frame+1}")
    return im, title

ani = animation.FuncAnimation(fig, update, frames=12, interval=1000, blit=False)

plt.show()
````
